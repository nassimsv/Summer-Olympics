# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
df=pd.read_csv('summer.csv')
print(len(df['City'].unique()))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
import pandas as pd
df=pd.read_csv('summer.csv')
data=df[df['Medal']=='Gold']
data=data.groupby(['Sport']).count()["Medal"].to_frame()
data.sort_values(by="Medal",ascending=False,inplace=True)
print(data.head(1))
data.head().plot(kind='bar',color='orange',figsize=(10,10))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
import pandas as pd
df=pd.read_csv('summer.csv')
data=df.groupby(['Sport']).count()["Medal"].to_frame()
data.sort_values(by="Medal",ascending=False,inplace=True)
data.head().plot(kind='bar',color='black',figsize=(10,10))
print(data.head(1))
```

### 4. Which player has won most number of medals? (Top 5)

```python
import pandas as pd
df=pd.read_csv('summer.csv')
data=df.groupby(['Athlete']).count()["Medal"].to_frame()
data.sort_values(by="Medal",ascending=False,inplace=True)
data.head().plot(kind='bar',color='green',figsize=(10,10))
print(data.head(1))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
import pandas as pd
df=pd.read_csv('summer.csv')
data=df[df['Medal']=='Gold']
data=data.groupby(['Athlete']).count()["Medal"].to_frame()
data.sort_values(by="Medal",ascending=False,inplace=True)
data.head().plot(kind='bar',color='purple',figsize=(10,10))
print(data.head(1))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
import pandas as pd
df=pd.read_csv('summer.csv')
df=df[df['Medal']=='Gold']
df=df[df['Country']=='IND']
print(df['Year'].min())
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
import pandas as pd
df=pd.read_csv('summer.csv')
df.drop_duplicates(subset=['Event','Athlete'],inplace=True)
df=df.groupby(['Event']).count()["Medal"].to_frame()
df.sort_values(by="Medal",ascending=False,inplace=True)
df.head().plot(kind='bar',color='grey',figsize=(10,10))
print(df.head(1))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
import pandas as pd
df=pd.read_csv('summer.csv')
df=df[df['Medal']=='Gold']
data=df[df['Gender']=="Women"]
data=data.groupby(['Sport']).count()["Medal"].to_frame()
data.sort_values(by="Medal",ascending=False,inplace=True)
data.head().plot(kind='bar',color='maroon',figsize=(10,10))
print(data.head(1))
```
